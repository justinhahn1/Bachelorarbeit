## Table 2 — Replication of Kempf & Ruenzi (2008)
### Impact of Segment Ranks on Net-Inflows

**Data sources used:**
- `kempf&ruenzi.csv` — monthly fund returns and TNA (CRSP)
- `fund summary.kempf.csv` — annual expense ratios, turnover, family (mgmt_name), segment (lipper_obj_name)
- `F-F_Research_Data_Factors_daily.csv` — Fama-French 3 factors
- `F-F_Momentum_Factor_daily.csv` — Carhart momentum factor (UMD)

**Notes on deviations from the original paper:**
- FEES uses `exp_ratio` only (paper adds 1/7 × total loads, which is not in the dataset)
- Segment ranks are based on Lipper objective codes (paper uses ICDI objective codes)
- Standard errors use Driscoll-Kraay kernel (closest available to PCSE in `linearmodels`)

```python
# ── Additional imports ──────────────────────────────────────────────
import warnings
warnings.filterwarnings('ignore')
from statsmodels.regression.linear_model import OLS
from statsmodels.tools import add_constant
from linearmodels.panel import PooledOLS
```

```python
# ── 1. Load raw data ─────────────────────────────────────────────────

# Monthly fund returns & TNA
kr = pd.read_csv('data/kempf&ruenzi.csv', low_memory=False)
kr['mret'] = pd.to_numeric(kr['mret'], errors='coerce')
kr['mtna'] = pd.to_numeric(kr['mtna'], errors='coerce')
kr['caldt'] = pd.to_datetime(kr['caldt'], errors='coerce')
kr = kr.dropna(subset=['caldt', 'mret', 'mtna'])
kr = kr.sort_values(['crsp_fundno', 'caldt']).reset_index(drop=True)
kr['crsp_fundno'] = kr['crsp_fundno'].astype(str).str.zfill(6)

# Annual fund summary (fees, turnover, family, segment)
fs = pd.read_csv('data/fund summary.kempf.csv', low_memory=False)
fs = fs[fs['summary_period2'] == 'A'].copy()
fs['caldt'] = pd.to_datetime(fs['caldt'], errors='coerce')
fs['crsp_fundno'] = fs['crsp_fundno'].astype(str).str.zfill(6)
fs['year'] = fs['caldt'].dt.year
fs['exp_ratio']  = pd.to_numeric(fs['exp_ratio'],  errors='coerce')
fs['turn_ratio'] = pd.to_numeric(fs['turn_ratio'], errors='coerce')

# Fama-French 3 factors (daily)
ff3 = pd.read_csv('data/F-F_Research_Data_Factors_daily.csv',
                  skiprows=4, header=0, low_memory=False)
ff3.columns = ff3.columns.str.strip()
ff3 = ff3.rename(columns={'Unnamed: 0': 'date'})
ff3['date'] = pd.to_datetime(ff3['date'].astype(str).str.strip(),
                              format='%Y%m%d', errors='coerce')
ff3 = ff3.dropna(subset=['date'])
for c in ['Mkt-RF', 'SMB', 'HML', 'RF']:
    ff3[c] = pd.to_numeric(ff3[c], errors='coerce') / 100

# Momentum factor (daily)
mom = pd.read_csv('data/F-F_Momentum_Factor_daily.csv',
                  skiprows=13, header=0, low_memory=False)
mom.columns = mom.columns.str.strip()
mom = mom.rename(columns={'Unnamed: 0': 'date', 'Mom': 'UMD'})
mom['date'] = pd.to_datetime(mom['date'].astype(str).str.strip(),
                              format='%Y%m%d', errors='coerce')
mom = mom.dropna(subset=['date'])
mom['UMD'] = pd.to_numeric(mom['UMD'], errors='coerce') / 100

print(f"KR: {len(kr):,} rows | {kr['crsp_fundno'].nunique():,} funds | "
      f"{kr['caldt'].min().year}–{kr['caldt'].max().year}")
print(f"Fund summary: {len(fs):,} rows | {fs['crsp_fundno'].nunique():,} funds")
```

```python
# ── 2. Aggregate monthly → annual ────────────────────────────────────

kr['year'] = kr['caldt'].dt.year

def compound(r):
    return (1 + r).prod() - 1

annual = (kr.groupby(['crsp_fundno', 'year'])
            .agg(tna      = ('mtna', 'last'),
                 ann_ret  = ('mret', compound),
                 ret_std  = ('mret', 'std'),
                 n_months = ('mret', 'count'))
            .reset_index())

# Require ≥10 months of data per year
annual = annual[annual['n_months'] >= 10].copy()

# ── 3. Merge fund summary ────────────────────────────────────────────

fs_merge = fs[['crsp_fundno', 'year', 'exp_ratio', 'turn_ratio',
               'mgmt_name', 'lipper_obj_name']].copy()
annual = annual.merge(fs_merge, on=['crsp_fundno', 'year'], how='left')

annual['fees']    = annual['exp_ratio']   # proxy: exp_ratio only
annual['segment'] = annual['lipper_obj_name'].fillna('Unknown')
annual['family']  = annual['mgmt_name'].fillna('Unknown_' + annual['crsp_fundno'])

# ── 4. Fund age ──────────────────────────────────────────────────────

first_year = (annual.groupby('crsp_fundno')['year']
                    .min().rename('first_year').reset_index())
annual = annual.merge(first_year, on='crsp_fundno', how='left')
annual['age'] = annual['year'] - annual['first_year'] + 1

# ── 5. Fund flow ─────────────────────────────────────────────────────
# Flow = (TNA_t - TNA_{t-1} * (1 + R_t)) / TNA_{t-1}

annual = annual.sort_values(['crsp_fundno', 'year'])
annual['tna_lag'] = annual.groupby('crsp_fundno')['tna'].shift(1)
annual['flow'] = ((annual['tna'] - annual['tna_lag'] * (1 + annual['ann_ret']))
                  / annual['tna_lag'])

# ── 6. Lagged controls ───────────────────────────────────────────────

annual['lnTNA'] = np.log(annual['tna'].clip(lower=0.001))
annual['lnAGE'] = np.log(annual['age'].clip(lower=1))

for col in ['lnTNA', 'lnAGE', 'ret_std', 'fees', 'turn_ratio', 'flow']:
    annual[f'{col}_lag'] = annual.groupby('crsp_fundno')[col].shift(1)

print(f"Annual panel: {len(annual):,} obs | {annual['crsp_fundno'].nunique():,} funds")
print(f"Segments: {annual['segment'].nunique()} | Families: {annual['family'].nunique()}")
```

```python
# ── 7. Monthly Fama-French factors ───────────────────────────────────

ff3['ym'] = ff3['date'].dt.to_period('M')
mom['ym'] = mom['date'].dt.to_period('M')

def cpd(s): return (1 + s).prod() - 1

mff = (ff3.groupby('ym')
          .agg(rf     = ('RF',     cpd),
               mkt_rf = ('Mkt-RF', cpd),
               smb    = ('SMB',    cpd),
               hml    = ('HML',    cpd))
          .reset_index())

mumd = (mom.groupby('ym')['UMD'].apply(cpd).reset_index()
           .rename(columns={'UMD': 'umd'}))

mff = mff.merge(mumd, on='ym', how='left')

kr['ym'] = kr['caldt'].dt.to_period('M')
kr = kr.merge(mff, on='ym', how='left')
kr['excess_ret'] = kr['mret'] - kr['rf']

print("Monthly FF factors merged.")
```

```python
# ── 8. Rolling performance metrics (trailing 36 months) ──────────────
# Sharpe Ratio, 3-Factor Alpha, 4-Factor Alpha
# Computed for each fund-year using months up to December of year t-1

def compute_metrics(grp, window=36):
    grp = grp.sort_values('caldt').reset_index(drop=True)
    rows = []
    for yr in grp['year'].unique():
        hist = grp[grp['year'] < yr].tail(window).dropna(
            subset=['excess_ret', 'mkt_rf', 'smb', 'hml', 'umd'])
        n = len(hist)
        sharpe = alpha3 = alpha4 = np.nan

        if n >= 12:
            er = hist['excess_ret'].values
            sharpe = er.mean() / er.std() * np.sqrt(12) if er.std() > 1e-10 else np.nan
            try:
                X3 = add_constant(hist[['mkt_rf', 'smb', 'hml']].values)
                alpha3 = OLS(er, X3).fit().params[0] * 12
            except Exception:
                pass
            try:
                X4 = add_constant(hist[['mkt_rf', 'smb', 'hml', 'umd']].values)
                alpha4 = OLS(er, X4).fit().params[0] * 12
            except Exception:
                pass

        rows.append({'year': yr, 'sharpe': sharpe,
                     'alpha3f': alpha3, 'alpha4f': alpha4})
    return pd.DataFrame(rows)

print("Computing rolling performance metrics (this takes ~2 minutes)...")
perf_list = []
for fund, grp in kr.groupby('crsp_fundno'):
    res = compute_metrics(grp)
    res['crsp_fundno'] = fund
    perf_list.append(res)

perf = pd.concat(perf_list, ignore_index=True)
annual = annual.merge(perf, on=['crsp_fundno', 'year'], how='left')
print(f"Done. Non-null Sharpe: {annual['sharpe'].notna().sum():,}")
```

```python
# ── 9. Segment ranks (quintile within segment × year) ────────────────

def quintile_rank(s):
    """Rank into quintiles 1–5; NaN if fewer than 5 valid obs."""
    if s.dropna().__len__() < 5:
        return pd.Series(np.nan, index=s.index)
    return pd.qcut(s, q=5, labels=[1, 2, 3, 4, 5],
                   duplicates='drop').astype(float)

for metric, col in [('sharpe',  'segrank_sr'),
                    ('alpha3f', 'segrank_3f'),
                    ('alpha4f', 'segrank_4f')]:
    annual[col] = (annual.groupby(['segment', 'year'], group_keys=False)
                         [metric].transform(quintile_rank))

# Lag by 1 year
for col in ['segrank_sr', 'segrank_3f', 'segrank_4f']:
    annual[f'{col}_lag'] = annual.groupby('crsp_fundno')[col].shift(1)

# LOW = Q1, MID = Q2–Q4, TOP = Q5
for suf, col in [('sr', 'segrank_sr_lag'),
                 ('3f', 'segrank_3f_lag'),
                 ('4f', 'segrank_4f_lag')]:
    annual[f'LOW_{suf}'] = (annual[col] == 1).astype(float)
    annual[f'MID_{suf}'] = annual[col].isin([2, 3, 4]).astype(float)
    annual[f'TOP_{suf}'] = (annual[col] == 5).astype(float)
    mask = annual[col].isna()
    annual.loc[mask, [f'LOW_{suf}', f'MID_{suf}', f'TOP_{suf}']] = np.nan

# ── 10. Family-level variables ───────────────────────────────────────

# lnTNA(Fam): log of total family TNA net of fund i
fam_tna = annual.groupby(['family', 'year'])['tna'].transform('sum')
annual['lnTNA_fam'] = np.log((fam_tna - annual['tna']).clip(lower=0.001))
annual['lnTNA_fam_lag'] = annual.groupby('crsp_fundno')['lnTNA_fam'].shift(1)

# STARRATIO(Fam): share of family funds in top quintile (Q5) in t-1
annual['is_star'] = (annual['segrank_sr'] == 5).astype(float)
annual.loc[annual['segrank_sr'].isna(), 'is_star'] = np.nan
fam_n     = annual.groupby(['family', 'year'])['crsp_fundno'].transform('count')
fam_stars = annual.groupby(['family', 'year'])['is_star'].transform('sum')
annual['starratio_fam'] = fam_stars / fam_n
annual['starratio_fam_lag'] = annual.groupby('crsp_fundno')['starratio_fam'].shift(1)

# FLOW(Fam): average family flow net of fund i in year t
fam_flow_sum = annual.groupby(['family', 'year'])['flow'].transform('sum')
annual['flow_fam'] = (fam_flow_sum - annual['flow']) / (fam_n - 1).clip(lower=1)

# FLOW(Seg): average segment flow net of fund i in year t
seg_n        = annual.groupby(['segment', 'year'])['crsp_fundno'].transform('count')
seg_flow_sum = annual.groupby(['segment', 'year'])['flow'].transform('sum')
annual['flow_seg'] = (seg_flow_sum - annual['flow']) / (seg_n - 1).clip(lower=1)

# ── 11. Winsorize flow at 1% / 99% ───────────────────────────────────

def winsorize(s, lo=0.01, hi=0.99):
    lb, ub = s.quantile(lo), s.quantile(hi)
    return s.clip(lb, ub)

for col in ['flow', 'flow_lag', 'flow_fam', 'flow_seg']:
    annual[col] = winsorize(annual[col])

print("Segment ranks and family variables computed.")
```

```python
# ── 12. Panel regression: PooledOLS + time FE + Driscoll-Kraay SE ────
# (Driscoll-Kraay is the closest available approximation to PCSE)

CONTROLS = ['ret_std_lag', 'turn_ratio_lag',
            'lnTNA_lag', 'lnAGE_lag', 'fees_lag',
            'flow_lag', 'lnTNA_fam_lag', 'starratio_fam_lag',
            'flow_fam', 'flow_seg']

def run_model(df_in, rank_suf):
    low, mid, top = f'LOW_{rank_suf}', f'MID_{rank_suf}', f'TOP_{rank_suf}'
    regs = [low, mid, top] + CONTROLS
    sub  = df_in[['flow', 'crsp_fundno', 'year'] + regs].dropna()
    sub  = sub.set_index(['crsp_fundno', 'year'])

    # Time fixed effects via year dummies
    yr_dummies = pd.get_dummies(
        sub.index.get_level_values('year'), prefix='yr', drop_first=True
    ).astype(float)
    yr_dummies.index = sub.index

    X = pd.concat([sub[regs], yr_dummies], axis=1)
    y = sub['flow']

    mod = PooledOLS(y, X)
    try:
        res = mod.fit(cov_type='kernel')
    except Exception:
        res = mod.fit(cov_type='robust')
    print(f"  {rank_suf}: {res.nobs:,} obs")
    return res

print("Running panel regressions...")
results = {suf: run_model(annual, suf) for suf in ['sr', '3f', '4f']}
```

```python
# ── 13. Build and display Table 2 ────────────────────────────────────
#hello world 
def sig_stars(p):
    if p < 0.01:  return '***'
    if p < 0.05:  return '**'
    if p < 0.10:  return '*'
    return ''

def cell(res, var):
    if var not in res.params.index:
        return '—'
    c, p = res.params[var], res.pvalues[var]
    sign = '+' if c >= 0 else ''
    return f"{sign}{c:.4f}{sig_stars(p)}"

VAR_ROWS = [
    ('LOW(SegRank)',     'LOW_sr',           'LOW_3f',           'LOW_4f'),
    ('MID(SegRank)',     'MID_sr',           'MID_3f',           'MID_4f'),
    ('TOP(SegRank)',     'TOP_sr',           'TOP_3f',           'TOP_4f'),
    ('STD',             'ret_std_lag',       'ret_std_lag',       'ret_std_lag'),
    ('TO',              'turn_ratio_lag',    'turn_ratio_lag',    'turn_ratio_lag'),
    ('lnTNA',           'lnTNA_lag',         'lnTNA_lag',         'lnTNA_lag'),
    ('lnAGE',           'lnAGE_lag',         'lnAGE_lag',         'lnAGE_lag'),
    ('FEES',            'fees_lag',          'fees_lag',          'fees_lag'),
    ('FLOW',            'flow_lag',          'flow_lag',          'flow_lag'),
    ('lnTNA(Fam)',      'lnTNA_fam_lag',     'lnTNA_fam_lag',     'lnTNA_fam_lag'),
    ('STARRATIO(Fam)',  'starratio_fam_lag', 'starratio_fam_lag', 'starratio_fam_lag'),
    ('FLOW(Fam)',       'flow_fam',          'flow_fam',          'flow_fam'),
    ('FLOW(Seg)',       'flow_seg',          'flow_seg',          'flow_seg'),
]

rows = []
for label, v_sr, v_3f, v_4f in VAR_ROWS:
    rows.append({
        'Variable':       label,
        'Sharpe Ratio':   cell(results['sr'], v_sr),
        '3-Factor Alpha': cell(results['3f'], v_3f),
        '4-Factor Alpha': cell(results['4f'], v_4f),
    })

rows.append({
    'Variable':       'R\u00b2',
    'Sharpe Ratio':   f"{results['sr'].rsquared:.2%}",
    '3-Factor Alpha': f"{results['3f'].rsquared:.2%}",
    '4-Factor Alpha': f"{results['4f'].rsquared:.2%}",
})

table2 = pd.DataFrame(rows).set_index('Variable')
table2.columns.name = 'Segment Ranks Based on'

# Style the table
styled = (table2.style
    .set_caption('Table 2 — Impact of Segment Ranks on Net-Inflows')
    .set_table_styles([
        {'selector': 'caption',
         'props': [('font-size', '14px'), ('font-weight', 'bold'),
                   ('text-align', 'center'), ('padding-bottom', '8px')]},
        {'selector': 'th',
         'props': [('text-align', 'center'), ('border-bottom', '1px solid black')]},
        {'selector': 'td',
         'props': [('text-align', 'right'), ('padding', '3px 12px')]},
        {'selector': 'tr:last-child td',
         'props': [('border-top', '1px solid black')]},
    ])
)

print(f"N obs — Sharpe: {results['sr'].nobs:,} | "
      f"3F: {results['3f'].nobs:,} | 4F: {results['4f'].nobs:,}")
styled
```
