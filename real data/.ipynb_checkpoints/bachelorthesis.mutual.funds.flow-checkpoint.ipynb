```python
CRSP Mutual Fund Data Preparation

This notebook prepares the raw CRSP mutual fund data for the empirical analysis of my bachelor thesis on mutual fund performance and fund flows.

The goal of this first step is to:
1. Load the raw CRSP monthly returns, fund summary, and fund-portfolio map files.
2. Inspect the structure and columns of each dataset.
3. Create column overview tables.
4. Prepare the basis for constructing a portfolio-month dataset.
```

```python
# ============================================================
# 0. Imports & project setup
# ============================================================
import pandas as pd
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.dpi'] = 120

# ── Project parameters ───────────────────────────────────────
PROJECT_NAME   = "CRSP Analysis"
BASE_DIR       = Path(".")          # current folder = notebook location
DATA_DIR       = BASE_DIR           # CSVs live in the same folder
PROCESSED_DIR  = BASE_DIR / "processed"
OUTPUT_DIR     = BASE_DIR / "output"

# Create output folders if they don't exist
PROCESSED_DIR.mkdir(exist_ok=True)
OUTPUT_DIR.mkdir(exist_ok=True)

print(f"Project : {PROJECT_NAME}")
print(f"Base dir: {BASE_DIR.resolve()}")
print(f"Processed → {PROCESSED_DIR.resolve()}")
print(f"Output    → {OUTPUT_DIR.resolve()}")

# ── Locate the three CRSP CSV files ─────────────────────────
csv_files = sorted(DATA_DIR.glob("*.csv"))
print(f"\nCSV files found in {DATA_DIR.resolve()}:")
for f in csv_files:
    print(f"  {f.name}")

if len(csv_files) < 3:
    raise FileNotFoundError(
        f"Expected at least 3 CSV files in '{DATA_DIR.resolve()}', "
        f"found {len(csv_files)}. Check the DATA_DIR path."
    )

# Assign the first three CSVs; rename variables to match your actual files
FILE_PATH_1 = csv_files[0]
FILE_PATH_2 = csv_files[1]
FILE_PATH_3 = csv_files[2]

print(f"\nFile 1: {FILE_PATH_1.name}")
print(f"File 2: {FILE_PATH_2.name}")
print(f"File 3: {FILE_PATH_3.name}")

# ============================================================
# 1. Load the three CSV files
# ============================================================
df1 = pd.read_csv(FILE_PATH_1, low_memory=False)
df2 = pd.read_csv(FILE_PATH_2, low_memory=False)
df3 = pd.read_csv(FILE_PATH_3, low_memory=False)

# ── Shapes ───────────────────────────────────────────────────
print("\n── Shapes ──────────────────────────────────────────────")
print(f"  {FILE_PATH_1.name:40s}  {df1.shape}")
print(f"  {FILE_PATH_2.name:40s}  {df2.shape}")
print(f"  {FILE_PATH_3.name:40s}  {df3.shape}")

# ── First rows ───────────────────────────────────────────────
print(f"\n── Head: {FILE_PATH_1.name} ──")
display(df1.head())

print(f"\n── Head: {FILE_PATH_2.name} ──")
display(df2.head())

print(f"\n── Head: {FILE_PATH_3.name} ──")
display(df3.head())

# ============================================================
# 2. Column overview tables
# ============================================================
def column_overview(df: pd.DataFrame, label: str) -> pd.DataFrame:
    """Return a tidy summary table for every column in *df*."""
    overview = pd.DataFrame({
        "column"    : df.columns,
        "dtype"     : df.dtypes.values,
        "non_null"  : df.notna().sum().values,
        "null"      : df.isna().sum().values,
        "null_pct"  : (df.isna().mean() * 100).round(2).values,
        "n_unique"  : df.nunique().values,
        "sample_val": [df[c].dropna().iloc[0] if df[c].notna().any() else np.nan
                       for c in df.columns],
    })
    overview.insert(0, "file", label)
    return overview.reset_index(drop=True)

overview1 = column_overview(df1, FILE_PATH_1.name)
overview2 = column_overview(df2, FILE_PATH_2.name)
overview3 = column_overview(df3, FILE_PATH_3.name)

# Display
print(f"\n── Column overview: {FILE_PATH_1.name} ──")
display(overview1)

print(f"\n── Column overview: {FILE_PATH_2.name} ──")
display(overview2)

print(f"\n── Column overview: {FILE_PATH_3.name} ──")
display(overview3)

# Save to processed folder
overview1.to_csv(PROCESSED_DIR / f"col_overview_{FILE_PATH_1.stem}.csv", index=False)
overview2.to_csv(PROCESSED_DIR / f"col_overview_{FILE_PATH_2.stem}.csv", index=False)
overview3.to_csv(PROCESSED_DIR / f"col_overview_{FILE_PATH_3.stem}.csv", index=False)

print("\nColumn overview tables saved to:", PROCESSED_DIR.resolve())
```
